```python
import os
import instructor
from pydantic import BaseModel
from google import genai
from dotenv import load_dotenv

# Load the variables from .env into the environment
load_dotenv()

# Define your structure
class UserInfo(BaseModel):
    name: str
    age: int

# Initialize client
# GenAI client will now find the key automatically from the environment
native_client = genai.Client(api_key=os.environ.get("GOOGLE_API_KEY"))
client = instructor.from_genai(native_client)

# Execute extraction
user = client.chat.completions.create(
    model="gemini-2.5-flash",
    response_model=UserInfo,
    messages=[
        {"role": "user", "content": "Extract: Alice is 25."}
    ],
)

print(user)
```

```python
import os
import instructor
from pydantic import BaseModel
from google import genai
from dotenv import load_dotenv

load_dotenv()

# 1. Define your Pydantic model
class BusinessInfo(BaseModel):
    name: str
    location: str
    revenue_estimate: int

# 2. Initialize the GenAI Client for Vertex AI
native_client = genai.Client(
    vertexai=True,
    project=os.environ.get("GOOGLE_CLOUD_PROJECT"),
    location=os.environ.get("GOOGLE_CLOUD_LOCATION")
)

# 3. Patch the client with Instructor
client = instructor.from_genai(native_client)

# 4. Create structured output
business = client.chat.completions.create(
    model="gemini-2.0-flash", 
    response_model=BusinessInfo,
    messages=[
        {"role": "user", "content": "The coffee shop 'Daily Grind' in Seattle"}
    ],
)

print(business)
```
